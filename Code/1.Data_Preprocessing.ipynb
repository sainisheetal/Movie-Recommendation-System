As a result, 13 columns have na values. To simplyfy, i am deleting columns which i am not going to use further.

```python
df1 = movies1.dropna()
```

```python
#delete useless columns
df = df1.drop(columns=['adult', 'belongs_to_collection', 'homepage', 'original_language', 'popularity', 
                      'status', 'tagline','video'])
```

```python
#Now there is no columns with null values but there are almost all the columns have some 0 values.
df.isna().sum()
```

```python


def genres_names(genres_list: str)-> str:
    genres_list = ast.literal_eval(genres_list)
    names = ""
    for genre_dict in genres_list:
            names+=(genre_dict['name'])+', '

    names=names[0:-2] #delete the last ', '
    return names


```

```python
#clean the 'genres' collumn
df['genres'] = df['genres'].apply(genres_names)
```

```python
#klib.clean - functions for cleaning datasets
klib.data_cleaning(df) # performs datacleaning (drop duplicates & empty rows/cols, adjust dtypes,...)
klib.clean_column_names(df) # cleans and standardizes column names, also called inside data_cleaning()
klib.convert_datatypes(df) # converts existing to more efficient dtypes, also called inside data_cleaning()
klib.drop_missing(df) # drops missing values, also called in data_cleaning()
klib.mv_col_handling(df) # drops features with high ratio of missing vals based on informational content
klib.pool_duplicate_subsets(df) # pools subset of cols based on duplicates with min. loss of information
```
